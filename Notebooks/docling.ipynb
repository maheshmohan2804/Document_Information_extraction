```python
# pip install docling python-dotenv groq
import os
from pathlib import Path
from docling.datamodel.base_models import InputFormat
from docling.datamodel.pipeline_options import PdfPipelineOptions, PictureDescriptionApiOptions
from docling.document_converter import DocumentConverter, PdfFormatOption

GROQ_API_KEY =os.environ["GROQ_KEY"]  # set this in your env

source = r"C:\Users\madha\OneDrive\Documents\PDF Data Extraction Pipeline\Data\s12887-020-02231-5.pdf"

# 1) Configure Docling to call a remote VLM
opts = PdfPipelineOptions(
    enable_remote_services=True,         # REQUIRED for remote APIs
)
opts.do_picture_description = True
opts.picture_description_options = PictureDescriptionApiOptions(
    # Groq’s OpenAI-compatible chat.completions endpoint:
    url="https://api.groq.com/openai/v1/chat/completions",
    headers={"Authorization": f"Bearer {GROQ_API_KEY}"},
    params={
        # Pick a Groq vision model (both accept image inputs):
        "model": "meta-llama/llama-4-scout-17b-16e-instruct",
        # optional tuning:
        "max_completion_tokens": 350,
        "temperature": 0.1,
        "top_p": 0.10,
        # seed=42  # (optional) for reproducibility
    },
    prompt="Describe the figure concisely and accurately, including axes/units if visible. If the image is a flowchart, describe the steps in order. If the image is a diagram, describe the components and their relationships. If the image is a graph, identify and describe the axes, trends and patterns.",
    timeout=90,
)

converter = DocumentConverter(format_options={
    InputFormat.PDF: PdfFormatOption(pipeline_options=opts)
})

doc = converter.convert(source).document

# 2) Read back the captions/descriptions
for page_idx, page in enumerate(doc.pages, start=1):
    # Many Docling builds expose page.pictures; if not, fall back to filtering all items.
    pics = getattr(page, "pictures", None)
    if pics is None:
        items = getattr(page, "items", [])
        pics = [it for it in items if it.__class__.__name__ == "PictureItem"]

    for pic in pics:
        print(f"Figure: {pic.self_ref} on page {page_idx}")
        # Human caption (from the PDF, if present)
        try:
            print("Caption:", pic.caption_text(doc=doc))
        except Exception:
            pass

        # AI description from your Groq VLM hook (if you enabled do_picture_description)
        for ann in getattr(pic, "annotations", []):
            if isinstance(ann, PictureDescriptionData):
                print("AI description:", ann.text)
        print("-" * 40)
```

```python
# Advanced Document Extraction with Chunking, Tagging, and Vector Storage
import os
import re
from groq import Groq
from typing import List, Dict
import chromadb
from chromadb.utils import embedding_functions

# Initialize clients
groq_client = Groq(api_key=os.environ["GROQ_KEY"])
chroma_client = chromadb.Client()

# Create or get collection
collection = chroma_client.get_or_create_collection(
    name="research_papers",
    embedding_function=embedding_functions.DefaultEmbeddingFunction()
)

print("Step 1: Chunking document by markdown headers...")
print("=" * 80)

# Get the markdown content
markdown_content = doc.export_to_markdown()

# Chunk by markdown headers (##, ###, etc.)
def chunk_by_markdown_headers(markdown_text: str) -> List[Dict[str, str]]:
    """Split markdown into chunks based on headers, keeping header with its content."""
    chunks = []
    
    # Split by headers (## or ###)
    # Pattern matches ## or ### followed by text, and captures everything until next header
    pattern = r'(^#{2,3}\s+.+?)(?=\n#{2,3}\s+|\Z)'
    matches = re.finditer(pattern, markdown_text, re.MULTILINE | re.DOTALL)
    
    for match in matches:
        chunk_text = match.group(1).strip()
        if chunk_text and len(chunk_text) > 50:  # Filter out very small chunks
            # Extract header
            header_match = re.match(r'^(#{2,3}\s+)(.+)', chunk_text)
            header = header_match.group(2).strip() if header_match else "Unknown"
            
            chunks.append({
                "header": header,
                "content": chunk_text,
                "char_count": len(chunk_text)
            })
    
    return chunks

chunks = chunk_by_markdown_headers(markdown_content)
print(f"Created {len(chunks)} chunks from the document\n")

# Display first few chunks
for i, chunk in enumerate(chunks[:5]):
    print(f"Chunk {i+1}: {chunk['header'][:60]}... ({chunk['char_count']} chars)")
print(f"... and {len(chunks) - 5} more chunks\n")
```

```python
print("Step 2: Tagging chunks with LLM and extracting metadata...")
print("=" * 80)

# Tagging prompt
tagging_prompt = """Analyze the following text chunk from a research paper and:

1. Assign relevant tags (return as comma-separated list):
   - <summary>: if chunk contains important details for paper summary (abstract, introduction, conclusion)
   - <research_methods>: if chunk describes research methodology, study design, data collection
   - <findings_conclusion>: if chunk contains results, findings, or conclusions
   - <metadata>: if chunk contains author names, publication date, affiliations

2. Extract metadata if present:
   - Authors: (list if found, otherwise "None")
   - Date: (date if found, otherwise "None")

Return your response in this exact format:
TAGS: <tag1>, <tag2>, ...
AUTHORS: author names or None
DATE: date or None

Text chunk:
{chunk_text}
"""

# Process chunks: tag and extract metadata
document_metadata = {"authors": None, "date": None}
tagged_chunks = []

for i, chunk in enumerate(chunks):
    print(f"Processing chunk {i+1}/{len(chunks)}: {chunk['header'][:50]}...")
    
    # Create tagging prompt
    prompt = tagging_prompt.format(chunk_text=chunk['content'][:2000])  # Limit to 2000 chars per chunk
    
    try:
        response = groq_client.chat.completions.create(
            model="llama-3.3-70b-versatile",
            messages=[{"role": "user", "content": prompt}],
            temperature=0.1,
            max_completion_tokens=200,
        )
        
        result = response.choices[0].message.content
        
        # Parse response
        tags = []
        authors = None
        date = None
        
        for line in result.split('\n'):
            if line.startswith('TAGS:'):
                tags_str = line.replace('TAGS:', '').strip()
                tags = [t.strip() for t in tags_str.split(',') if t.strip()]
            elif line.startswith('AUTHORS:'):
                authors_str = line.replace('AUTHORS:', '').strip()
                if authors_str and authors_str.lower() != 'none':
                    authors = authors_str
                    if not document_metadata["authors"]:
                        document_metadata["authors"] = authors
            elif line.startswith('DATE:'):
                date_str = line.replace('DATE:', '').strip()
                if date_str and date_str.lower() != 'none':
                    date = date_str
                    if not document_metadata["date"]:
                        document_metadata["date"] = date
        
        # Store tagged chunk
        tagged_chunks.append({
            **chunk,
            "tags": tags,
            "chunk_id": i
        })
        
    except Exception as e:
        print(f"  Error: {e}")
        tagged_chunks.append({
            **chunk,
            "tags": [],
            "chunk_id": i
        })

print(f"\nExtracted Metadata:")
print(f"  Authors: {document_metadata['authors']}")
print(f"  Date: {document_metadata['date']}")
print(f"\nTagged {len(tagged_chunks)} chunks\n")
```

```python
print("Step 3: Creating embeddings and storing in ChromaDB...")
print("=" * 80)

# Prepare data for ChromaDB
documents = []
metadatas = []
ids = []

for chunk in tagged_chunks:
    documents.append(chunk['content'])
    metadatas.append({
        "header": chunk['header'],
        "tags": ','.join(chunk['tags']),
        "char_count": chunk['char_count'],
        "chunk_id": chunk['chunk_id']
    })
    ids.append(f"chunk_{chunk['chunk_id']}")

# Add to ChromaDB collection
collection.add(
    documents=documents,
    metadatas=metadatas,
    ids=ids
)

print(f"Stored {len(documents)} chunks in ChromaDB with embeddings")
print(f"Collection now has {collection.count()} items\n")

# Show tag distribution
tag_counts = {}
for chunk in tagged_chunks:
    for tag in chunk['tags']:
        tag_counts[tag] = tag_counts.get(tag, 0) + 1

print("Tag distribution:")
for tag, count in sorted(tag_counts.items(), key=lambda x: x[1], reverse=True):
    print(f"  {tag}: {count} chunks")
```

```python
print("Step 4: Implementing Hybrid Retrieval (Semantic + BM25)...")
print("=" * 80)

from rank_bm25 import BM25Okapi
import numpy as np

# Prepare BM25 index
tokenized_docs = [doc.lower().split() for doc in documents]
bm25 = BM25Okapi(tokenized_docs)

def hybrid_search(query: str, tag_filter: str = None, top_k: int = 5, alpha: float = 0.5):
    """
    Hybrid search combining semantic similarity and BM25 keyword matching.
    
    Args:
        query: Search query
        tag_filter: Filter by tag (e.g., '<summary>', '<research_methods>')
        top_k: Number of results to return
        alpha: Weight for semantic search (1-alpha for BM25). 0.5 = equal weight
    
    Returns:
        List of relevant chunks with scores
    """
    # 1. Semantic search using ChromaDB
    where_filter = {"tags": {"$contains": tag_filter}} if tag_filter else None
    
    semantic_results = collection.query(
        query_texts=[query],
        n_results=top_k * 2,  # Get more for reranking
        where=where_filter
    )
    
    # 2. BM25 keyword search
    tokenized_query = query.lower().split()
    bm25_scores = bm25.get_scores(tokenized_query)
    
    # 3. Combine scores (normalize and weight)
    # Get semantic scores (convert distances to similarities)
    semantic_ids = semantic_results['ids'][0]
    semantic_distances = semantic_results['distances'][0]
    
    # Normalize scores to 0-1 range
    if len(semantic_distances) > 0:
        max_dist = max(semantic_distances) if max(semantic_distances) > 0 else 1
        semantic_scores_norm = {
            semantic_ids[i]: 1 - (semantic_distances[i] / max_dist)
            for i in range(len(semantic_ids))
        }
    else:
        semantic_scores_norm = {}
    
    # Normalize BM25 scores
    max_bm25 = max(bm25_scores) if max(bm25_scores) > 0 else 1
    bm25_scores_norm = bm25_scores / max_bm25
    
    # Combine scores
    combined_scores = {}
    for i, chunk_id in enumerate(ids):
        sem_score = semantic_scores_norm.get(chunk_id, 0)
        bm25_score = bm25_scores_norm[i]
        
        # Weighted combination
        combined_scores[chunk_id] = alpha * sem_score + (1 - alpha) * bm25_score
    
    # Sort by combined score
    top_chunks = sorted(combined_scores.items(), key=lambda x: x[1], reverse=True)[:top_k]
    
    # Return chunk details
    results = []
    for chunk_id, score in top_chunks:
        chunk_idx = int(chunk_id.split('_')[1])
        chunk = tagged_chunks[chunk_idx]
        results.append({
            "chunk_id": chunk_id,
            "score": score,
            "header": chunk['header'],
            "content": chunk['content'][:500] + "..." if len(chunk['content']) > 500 else chunk['content'],
            "tags": chunk['tags']
        })
    
    return results

# Test hybrid search
print("Testing hybrid search with query: 'What are the research methods?'\n")
results = hybrid_search("What are the research methods?", tag_filter="<research_methods>", top_k=3)

for i, result in enumerate(results, 1):
    print(f"{i}. {result['header']} (score: {result['score']:.3f})")
    print(f"   Tags: {', '.join(result['tags'])}")
    print(f"   Preview: {result['content'][:200]}...")
    print()
```

```python
print("Step 5: Generate Final Document Extraction using RAG...")
print("=" * 80)

# Retrieve relevant chunks for each section
summary_chunks = hybrid_search(
    "abstract introduction background summary overview", 
    tag_filter="<summary>", 
    top_k=3,
    alpha=0.6  # Favor semantic search
)

methods_chunks = hybrid_search(
    "research methods methodology study design data collection analysis approach",
    tag_filter="<research_methods>",
    top_k=3,
    alpha=0.4  # Favor keyword search for methods
)

findings_chunks = hybrid_search(
    "results findings conclusions discussion outcomes implications",
    tag_filter="<findings_conclusion>",
    top_k=3,
    alpha=0.5  # Balanced
)

# Compile context for final extraction
def compile_context(chunks_list):
    """Combine chunk contents into a single context string."""
    return "\n\n---\n\n".join([chunk['content'] for chunk in chunks_list])

summary_context = compile_context(summary_chunks)
methods_context = compile_context(methods_chunks)
findings_context = compile_context(findings_chunks)

# Final extraction prompt
final_extraction_prompt = """Based on the provided document chunks, extract the following information:

## Document Information
**Author(s):** {authors}
**Publication Date:** {date}

## Document Summary
[Provide a 2-3 sentence summary based on the context below]

## Research Methods
[Summarize the research methodology including:
- Study design
- Data sources
- Sample size and population
- Analytical methods]

## Key Findings and Conclusions
[Summarize main findings and conclusions including:
- Primary outcomes
- Statistical significance
- Main conclusions
- Implications]

---

SUMMARY CONTEXT:
{summary_context}

---

METHODS CONTEXT:
{methods_context}

---

FINDINGS CONTEXT:
{findings_context}
"""

# Format the prompt
formatted_prompt = final_extraction_prompt.format(
    authors=document_metadata['authors'] or "Not found",
    date=document_metadata['date'] or "Not found",
    summary_context=summary_context[:3000],  # Limit context size
    methods_context=methods_context[:3000],
    findings_context=findings_context[:3000]
)

print("Generating final extraction with Groq...\n")

# Generate final extraction
response = groq_client.chat.completions.create(
    model="llama-3.3-70b-versatile",
    messages=[{"role": "user", "content": formatted_prompt}],
    temperature=0.2,
    max_completion_tokens=1500,
)

final_extraction = response.choices[0].message.content

print("=" * 80)
print("FINAL DOCUMENT EXTRACTION")
print("=" * 80)
print(final_extraction)
print("=" * 80)
```
